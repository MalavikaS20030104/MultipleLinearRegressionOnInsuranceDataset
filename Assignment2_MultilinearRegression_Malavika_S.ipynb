```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
df=pd.read_csv("C:/Users/malav/Downloads/Concrete_Data - Concrete_Data.csv")
df.head(8)
```

```python
X=df[["Cement","BlastFurnaceSlag","FlyAsh","Water","CoarseAggregate","FineAggregate","AgeInDays",]]
Y=df["ConcreteCompressiveStrength"]
```

```python
x_train,x_test,y_train,y_test=train_test_split(X,Y,test_size=0.2,random_state=2)
model=LinearRegression()
model.fit(x_train,y_train)
y_pred=model.predict(x_test)
y_pred
```

```python
a=[[300,140,0,190,1000,663,11]]
a_pred=model.predict(a)
print(a_pred)
```

```python
from sklearn.metrics import mean_squared_error,root_mean_squared_error,mean_absolute_error,mean_absolute_percentage_error
mse=mean_squared_error(y_test,y_pred)
print("mse:",mse)
```

```python
mae=mean_absolute_error(y_test,y_pred)
print("mae:",mae)
```

```python
mape=mean_absolute_percentage_error(y_test,y_pred)
print("mape:",mape)
```

```python
rmse=root_mean_squared_error(y_test,y_pred)
print("rmse",rmse)
```

```python

import matplotlib.pyplot as plt
plt.scatter(y_test,y_pred, color='green')
plt.title("predicted ConcreteCompressiveStrength v/s actual ConcreteCompressiveStrength")
plt.xlabel("actual ConcreteCompressiveStrength")
plt.ylabel("predicted ConcreteCompressiveStrength")
plt.legend()
plt.show()
```

matplotlib  is one of the major libraries used in machine learning inorder to plot the relationship between two features across the instances present in the data provided.It gives a scattered pattern by plotting the actual target values and predicted target values.

```python
import seaborn as sns
corr_matrix=df.corr()
plt.figure(figsize=(10,8))
sns.heatmap(corr_matrix,annot=True,cmap='coolwarm',vmin=-1,vmax=1)
plt.title("heatmap")
plt.show()
```

Heatmap is data representation where the variations in data is represented by colours. For developing heatmap, we find teh correlation matrix of a dataframe.Correlation matrix is a matrix where each cell represents the relationship between two variables.After that we call the instance for heatmap() in which we pass the correlation matrix. Along with the correlation matrix, we also pass certain parameters which plays a keyrole in the development of heatmap.they are:
 1) annot:True- setting the annot parameter to true value shows the correlation value
 2) cmap= 'coolwarm' gives the diverse asthetics
 3) vmin=-1, vmax=1 make sure the values stay under (-1,1)

```python
sns.histplot(data=df,x='ConcreteCompressiveStrength',kde=True)
plt.title('histogram')
plt.xlabel('ConcreteCompressiveStrength values')
plt.ylabel('frequency')
plt.show()
```

```python
residuals=y_test-y_pred
plt.scatter(y_pred, residuals)
plt.axhline(y=0, color='r', linestyle='--')
plt.xlabel("Predicted Values")
plt.ylabel("Residuals")
plt.title("Residual Plot")
plt.show()
```

Residual plot- residual plot is the typw plot which give errors on y axis and the predicted values.The difference between actual target values and predicted values is known as the residual value.Now an axis line, parallel to y-axis is plotted to analyse the distribution of data.
